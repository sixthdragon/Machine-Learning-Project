## Function To Get Page Numbers Of The Product

```python
templates=[]
def get_url(search_term):
    for i in range(1,29):
        template="https://www.flipkart.com/search?q={}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}"
        search_term=search_term.replace(' ',"%20")
        template=template.format(search_term,i)
        templates.append(template)
        
```

```python
#can be used to search any product on flipkart
url=get_url("air conditioners")
len(templates)
```

28 are number of pages present in flipkart for product Air Conditioners

## Code for getting names of all products present on the Flipkart

```python
products=[]
for url in templates:
    r=requests.get(url)
    htmlContent=r.content
    soup=BeautifulSoup(htmlContent,"html.parser")
    #soup=BeautifulSoup(driver.page_source,'html.parser')
    #content=driver.page_source.encode("utf-8").strip()
    #soup=BeautifulSoup(content,"html.parser")
    for a in soup.findAll('a',href=True,attrs={'class':'_1fQZEK'}):
        try :
            name=a.find('div',attrs={'class':'_4rR01T'})
            products.append(name.text)
        except:
            products.append("NA")
```

# Script For Data Extraction Of Products

```python
# table data extraction
import time
from selenium.common.exceptions import StaleElementReferenceException
import requests
import bs4
for x in products_list:
    print("\n\n")
    try:
        #enter the values in searchbar
        search = driver.find_element_by_name("q")
        search.clear()
        search.send_keys(x)
        search.send_keys(Keys.RETURN)
        print(products_list.index(x)," ", x)
        
    except StaleElementReferenceException as Exception:
        
        search = driver.find_element_by_name("q")
        search.clear()
        search.send_keys(x)
        search.send_keys(Keys.RETURN)
        print(products_list.index(x)," ",x)
    try:   
        #get the url for the 1st item
        for a in driver.find_elements_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[2]/div/div/div/a'):
            url = a.get_attribute("href")
    except StaleElementReferenceException as Exception:
        for a in driver.find_elements_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[2]/div/div/div/a'):
            url = a.get_attribute("href")
    print(url)
    try:
        res = requests.get(url)
        res.raise_for_status()
        soup = bs4.BeautifulSoup(res.text, "html.parser")
        price = soup.find(class_='_30jeq3 _16Jk6d')
        p = price.text.strip()
        accuracy['Price'][accuracy.index == x] = p[1:]
        while(True):
            table = soup.find("table", class_="_14cfVK")
            row_data = []
            key = []
            value = []
            for row in table.findAll("tr"):
                cols = row.find_all("td")
                k = cols[0].text.strip()
                if k in accuracy.columns:
                    key.append(k)
                    for ele in cols[1:]:
                        v = ele.text.strip()
                    value.append(v)
                    accuracy[k][accuracy.index == x] = v
            
            row_data = { k:v for (k,v) in zip(key, value)}  
            print(row_data)
            if table is None:
                break
            table.decompose()       ##drop the table
    except:
        continue
```

We have Used Selenium For Navigating through pages by taking the xpath from the website and interacting with the search bar and searching each and every product from the search bar, then by the help of beautiful soup we have parsed the page and collected information about the specifications table in a class name by call tr and td tag and the price too.
